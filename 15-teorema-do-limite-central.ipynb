# Teorema do Limite Central

<font size="3">Independentemente da distribuição da população, a distribuição das médias de amostras aleatórias tende a se aproximar de uma distribuição normal à medida que o tamanho das amostras aumenta.</font>

Normal = N($\mu$, $\sigma$)

A média da distribuição das médias é $\Large \mu_{\bar{x}} = \mu$

Para população **finita amostra sem reposição**:

O desvio padrão da distribuição das médias é $\Large \sigma_{\bar{x}} = \frac{\sigma}{\sqrt{n}}\sqrt{\frac{N - n}{N -1}}$

Para população **infinita com reposição**:

O desvio padrão da distribuição das médias é $\Large \sigma_{\bar{x}} = \frac{\sigma}{\sqrt{n}}$

Exemplo usando Python jogar dados

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statistics as stats

np.random.seed(123)
```

```python
omega = (1, 2, 3, 4, 5, 6)
dado = np.array(omega)
dado
```

### Média

```python
print('Média = ', dado.mean())
```

Jogar Aleatóriamente

```python
np.random.randint(1, 7)
```

### **20** Médias amostrais jogando o dado **40** vezes

```python
medias_amostrais_1 = [np.mean(np.random.randint(1,7,40)) for _ in range(20)]
plt.hist(medias_amostrais_1)
plt.title('20 Médias conjutos com 40 eventos')
plt.show()
```

### 1000 Médias amostrais jogando o dado 40 vezes

```python
medias_amostrais_2 = [np.mean(np.random.randint(1,7,40)) for _ in range(1000)]
plt.hist(medias_amostrais_2)
plt.title('1000 Médias conjutos com 40 eventos')
plt.show()
```

## **Média** de **1000** Médias amostrais jogando o dado **40** vezes

```python
print('Média = ', stats.mean(medias_amostrais_2))
```

### **Mediana** de **1000** Médias amostrais jogando o dado **40** vezes

```python
print('Mediana= ', stats.median(medias_amostrais_2))
```

### **Moda** de **1000** Médias amostrais jogando o dado **40** vezes

```python
print('Moda = ', stats.mode(medias_amostrais_2))
```

## Teorema do Limite Central em um DataFrame
Aplicando o mesmo conceito com um DataFrame de um arquivo CSV

```python
df = pd.read_csv('data/altura-peso-br.csv')
df.head()
```

### Média da População

```python
stats.mean(df['Peso(kg)'])
```

### Gerar amostra aleatória 5 elementos

```python
df['Peso(kg)'].sample(n=5)
```

### Gerar 200 médias amostrais de 50 amostras cada

```python
np.random.seed(123)
medias_amostrais = []
for _ in range(200):
    amostras = df['Peso(kg)'].sample(n=50)
    media_amostral = np.mean(amostras)
    medias_amostrais.append(media_amostral)
```

```python
len(medias_amostrais)
```

```python
plt.hist(medias_amostrais)
plt.show()
```

```python
print('Média das Médias = ', stats.mean(medias_amostrais))
```
