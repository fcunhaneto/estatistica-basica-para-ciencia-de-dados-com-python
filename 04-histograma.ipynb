```python
import pandas as pd
import matplotlib.pyplot as plt
import create_frequency_table as ctf
```

## Trabalhadores dist√¢ncias de casa para o trabalho

```python
# Trabalhadores dist√¢ncias de casa para o trabalho
km = pd.Series([4,6,6,7,11,13,18,19,21,24,26,27,35,36,36,42,43,45,49])
km.plot.hist()
plt.show()
```

## Escolhendo o n√∫mero de barras

```python
barras = 5
km.plot.hist(bins=barras)
plt.show()
```

```python
## C√°lculo para fazer a divis√£o do n√∫mero em barras
```

### Obtendo o m√≠nimo e o m√°ximo valor

```python
primeiro = km.min()
ultimo = km.max()

print(primeiro, ultimo)
```

### Obtendo o tamanho do intervalo (ultimo - primeiro)/ numero de barras = (49-5)/5

```python
print('Tamanho do intervalo = ', (ultimo - primeiro)/5)
```

#### Definido os n√∫meros que v√£o no eixo x

```python
ticks = ctf.ticks(4, 49, amplitude=9)
print(ticks)
```

```python
km.plot.hist(bins=barras)

plt.xticks(ticks)
plt.title('Dist√¢ncia de Casa para o Trabalho')
plt.xlabel('Km')
plt.ylabel('Frequ√™ncia')

plt.show()
```

#### Dando uma dist√¢ncia entre as barras

```python
km.plot.hist(bins=barras, rwidth=0.95)

plt.xticks(ticks)
plt.title('Dist√¢ncia de Casa para o Trabalho')
plt.xlabel('Km')
plt.ylabel('Frequ√™ncia')
plt.grid(axis='y')

plt.show()
```
