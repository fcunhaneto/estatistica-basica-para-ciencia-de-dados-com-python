```python
import pandas as pd
import statistics as st
import create_frequency_hist as cfh
```

```python
df = pd.read_csv('data/altura-peso-br.csv')
df.head()
```

```python
mh = df.loc[df.Sexo == 'M']['Altura(m)']
mh.head(7)
```

```python
cfh.freq_hist(data_series=mh, bars=9, title='Altura Homens', unit='Metros')
```

## Média Altura Homens

```python
print('Média = ', st.mean(mh))
```

## Desvio Padrão Homens

```python
print('S = ', st.stdev(mh))
```

Dentro dessa população qual a probabilidade de encontrar alguem com menos de 1,65 metros?

```python
Z = (165 - st.mean(mh)) / (st.stdev(mh))
print('Z = ', round(Z, 2))
```

Para Z = 1.42

```python
P = 0.5 - 0.4573
print('P = {0:.2f}%'.format(round(P, 2)*100))
```
